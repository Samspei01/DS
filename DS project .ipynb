Engine Size: The size of the engine in liters.
Body Type: The type of car body.
MPG: The miles per gallon the car gets.
Road Tax: The amount of road tax payable for the car in Euros.

# Data Cleaning:

Dropping the unnecessary columns:

```python
##mercedes_df = mercedes_df.drop(["ID", "Model"], axis=1)
```

Renaming the columns for better understanding:

```python
mercedes_df = mercedes_df.rename(columns={"Year": "Manufacturing Year", "Fuel Type": "Fuel_Type"})
```

Finding and removing the duplicate values:

```python
mercedes_df.drop_duplicates(inplace=True)
```

Handling missing data by removing the rows containing missing values:

```python
mercedes_df.dropna(inplace=True)
```

Creating a new column for the car's age:

```python
mercedes_df["Car Age"] = 2023 - mercedes_df["year"]
```

Extracting the numerical values from the Mileage column:

```python
print(mercedes_df["mileage"].dtype)
```

```python
mercedes_df["mileage"] = mercedes_df["mileage"].astype(str)
```

```python
print(mercedes_df["mileage"].dtype)
```

Then,apply the str accessor to the "mileage" column and remove the " miles" substring as follows:



```python
mercedes_df["mileage"] = mercedes_df["mileage"].str.replace(" miles", "")
```

After that,convert the "mileage" column to numeric using the pd.to_numeric method as follows:

```python
mercedes_df["mileage"] = pd.to_numeric(mercedes_df["mileage"])
```

Remove duplicates: Check for any duplicate rows in the data and remove them if they exist.

```python
mercedes_df = mercedes_df.drop_duplicates()
```

Check for missing values: Check if any columns have missing values and decide how to handle them. You can either remove rows with missing values or fill in the missing values with appropriate values.

```python
valid_columns = ['price', 'year', 'mileage', 'tax', 'mpg', 'engineSize']

# Fill missing values in selected columns with the mean
mercedes_df[valid_columns] = mercedes_df[valid_columns].fillna(mercedes_df[valid_columns].mean())

# Check for missing values
missing_values = mercedes_df.isnull().sum()
print(missing_values)

# Drop rows with missing values
mercedes_df.dropna(inplace=True)
```

# Data Visualization:

 Plotting a scatter plot between Price and Mileage:

```python
sns.scatterplot(x="mileage", y="price", data=mercedes_df)
plt.title("Scatter Plot of Price vs Mileage")
plt.show()
```

 Plotting a bar chart showing the count of each Fuel Type:


```python
sns.countplot(x="fuelType", data=mercedes_df)
plt.title("Bar Chart of Fuel Type")
plt.show()
```

#### Box plot of price by fuel type:

This will show you how the price of Mercedes cars varies by fuel type, allowing you to compare the median, quartiles, and outliers.

```python
sns.boxplot(x="fuelType", y="price", data=mercedes_df)
plt.title("Box Plot of Price by Fuel Type")
plt.show()
```

Histogram of mileage:

This will show you the distribution of mileage in the dataset, allowing you to see how many cars fall within different mileage ranges.

```python
sns.histplot(x="mileage", data=mercedes_df, bins=20)
plt.title("Histogram of Mileage")
plt.show()
```

Scatter plot of price vs. engine size, colored by fuel type:

```python
sns.scatterplot(x="engineSize", y="price", hue="fuelType", data=mercedes_df)
plt.title("Scatter Plot of Price vs Engine Size, Colored by Fuel Type")
plt.show()
```

Heatmap of correlation matrix:

This will show you the correlation matrix of the numerical columns in the dataset, allowing you to see which variables are most strongly correlated with each other.

```python
# Select numeric columns
numeric_columns = mercedes_df.select_dtypes(include='number')

# Compute correlation matrix
corr = numeric_columns.corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```
